```python
import rasterio, matplotlib, scipy.ndimage
from rasterio.windows import Window
from rasterio.plot import show

# CONSTANTS
WORLDVIEW_IMAGE = '/home/andrew/data/imagery/61b94e69b26de1000596d607.tif'

# Read the raster data into a window
with rasterio.open(WORLDVIEW_IMAGE) as src:
    w = src.read([1,2,3], window=Window(800, 400, 500,500))
# Resample the windowed image
resampled = scipy.ndimage.zoom(w, [1,0.5,0.5])
# Change the size of the axes
fig, axes = matplotlib.pyplot.subplots(figsize=(14,14))
# axes.set_axis_off()
# Show the data
show(resampled, ax=axes)
```
